```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as s
import plotly.express as px

df=pd.read_csv(r"IPL.csv")
df.head()
```

```python
df.info()
df.isnull().sum()
df.describe()
```

```python
df.drop_duplicates(inplace=True)
df["city"].fillna("Unknown",inplace=True)
df["match_won_by"].fillna("No Result",inplace=True)

df["date"]=pd.to_datetime(df["date"])
```

```python
df["season"] = pd.to_numeric(df["season"], errors="coerce")
df = df.dropna(subset=["season"])
df["season"] = df["season"].astype(int)
matches_per_season = (
    df.groupby("season")["match_id"]
      .nunique()
      .sort_index()
)


plt.figure(figsize=(12,5))
plt.plot(matches_per_season.index,matches_per_season.values,marker="o")

plt.title("IPL Matches Played Per season (2004-2025)")
plt.xlabel("Season")
plt.ylabel("Number of Matches")
plt.grid(alpha=0.3)
plt.show()
```

```python
top_teams=(df.groupby("match_won_by")["match_id"]
           .nunique().sort_values(ascending=False).head(10))
```

```python
plt.figure(figsize=(9,5))
s.barplot(x=top_teams.values,y=top_teams.index,color="skyblue",edgecolor="blue")
plt.title("Top 10 Most Successful IPL Teams(2004-2025)")
plt.xlabel("Wins")
plt.ylabel("Teams")
plt.show()
```

```python
toss_advantage=df[df["toss_winner"]==df["match_won_by"]]
percentage=(len(toss_advantage)/len(df))*100
percentage
```

```python
venue_matches=(df.groupby("venue")["match_id"]
               .nunique().sort_values(ascending=False)
               .head(10))
venue_matches_df = venue_matches.reset_index(name='Matches')
venue_matches_df.rename(columns={'venue': 'Venue'}, inplace=True)

fig=px.bar(venue_matches_df, x="Matches", y="Venue", orientation="h",
           title="Top IPL Venues by Matches Hosted (2004-2025)",
           text="Matches")
fig.show()
```

```python
batting = (
    df.groupby("batter")
      .agg(
          batter_runs=("batter_runs", "sum"),
          balls_faced=("batter", "size")
      )
      .reset_index()
)

batting["strike_rate"] = (
    batting["batter_runs"] / batting["balls_faced"]

) * 100
top_batsmen = batting[
    (batting["batter_runs"] > 3000) &
    (batting["balls_faced"] > 2000)
]
top_10_strike_rate = (
    top_batsmen
    .sort_values(by="strike_rate", ascending=False)
    .head(10)
)

top_10_strike_rate

```

```python
top_batters_by_runs = (
    batting[batting["balls_faced"] > 2000]
    .sort_values(by="batter_runs", ascending=False)
    .head(10)
)

top_batters_by_runs
```

```python
plt.figure(figsize=(9,5))
s.barplot(
    x="batter_runs",
    y="batter",
    data=top_batters_by_runs,
    color="blue",
    edgecolor="blue",
    alpha=0.5
)

plt.title("Top 10 IPL Batters by Total Runs")
plt.xlabel("Total Runs")
plt.ylabel("Batter")
plt.show()
```

```python
batter_team = (
    df.groupby(["batter", "batting_team"])["batter_runs"]
      .sum()
      .reset_index()
)


primary_team = (
    batter_team
    .sort_values("batter_runs", ascending=False)
    .drop_duplicates("batter")
)

primary_team = primary_team[["batter", "batting_team"]]
```

```python
batting_teamwise = batting.merge(
    primary_team,
    on="batter",
    how="left"
)
filtered = batting_teamwise[
    (batting_teamwise["balls_faced"] > 2000) &
    (batting_teamwise["batter_runs"] > 3000)
]

fig = px.scatter(
    filtered,
    x="batter_runs",
    y="strike_rate",
    color="batting_team",
    size="balls_faced",
    hover_name="batter",
    title="IPL Batters: Runs vs Strike Rate (Team-wise)"
)

fig.show()

```
